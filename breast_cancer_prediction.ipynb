<h2>Kmeans Clustering</h2>


```python
class Scratch_KmeansClustering:

#Total clusters and max iter are the two parameters that the function takes.
#The "total clusters" and "max iter" attributes of the class are set using this method to the input values.
    def __init__(self, total_clusters, max_iter=100):
        self.total_clusters = total_clusters
        self.max_iter = max_iter


    def fit_predict(self, X):
          np.random.seed(0)
          centroids = X[np.random.choice(X.shape[0], size=self.total_clusters), : ] # Choose data points from X called "total clusters" at random to serve as the starting centroids.

          i = 0
          while i < self.max_iter:
              new_X = X.reshape(X.shape[0], 1, X.shape[1])
              new_centroids = np.expand_dims(centroids, axis=0)

              # Determine the distance between every data point and every centroid.
              distance = new_X - new_centroids
              sq_dist = distance ** 2
              dist = np.sum(sq_dist, axis=-1)

              # Determine the distances between every data point and every centroid.
              distances = np.zeros((X.shape[0], self.total_clusters))
              for j in range(self.total_clusters):
                  centroid = centroids[j]
                  distance = np.sum((X - centroid) ** 2, axis=1)
                  distances[:, j] = distance

              # Place each data point's centroid nearest to it.
              y_predicted = np.zeros(X.shape[0])
              for i in range(X.shape[0]):
                  shortest_distance = float("9e999")
                  for j in range(self.total_clusters):
                      dist = distances[i, j]
                      if dist < shortest_distance:
                          shortest_distance = dist
                          y_predicted[i] = j
              # Based on the provided data points, update the centroid locations.
              for j in range(self.total_clusters):
                  clustered_data_points = X[y_predicted == j]
                  average_point = np.mean(clustered_data_points, axis=0)
                  centroids[j] = average_point

              i= i+ 1
          # Provide each data point's anticipated cluster labels back.
          return y_predicted
```

<h2>Spectral Clustering</h2>

```python
class Scratch_SpectralClustering:

    #The "total clusters" attributes of the class are set using this method to the input values.
    def __init__(self, total_clusters):
        self.total_clusters = total_clusters

    def fit_predict(self, X):
        # Calculate the median distance between adjacent points in X.
        pair_dist = np.median(pairwise_distances(X, metric='euclidean'))
        dist = np.sum(X**2)
        mul = np.dot(X, X.T)
        sq_sum = np.sum(np.square(X), axis=1)
        squared_distances = sq_sum.reshape(-1, 1) + sq_sum - 2 * mul
        K = np.exp(-pair_dist * squared_distances)

       # Calculate the Laplacian matrix L and the degree matrix D.
        sums = np.sum(K, axis=1)
        D = np.zeros((K.shape[0], K.shape[0]))
        for i in range(K.shape[0]):
            D[i, i] = sums[i]
        L = D - K

        # Calculate L eigenvalues and eigenvectors
        eigenvalues, eigenvectors = np.linalg.eigh(L)

        # Choose your first eigenvectors from total clusters are stored in V.
        V = eigenvectors[:, :self.total_clusters]

        # Applying K-means clustering on V
        kmeans = Scratch_KmeansClustering(total_clusters=self.total_clusters)
        y_predicted = kmeans.fit_predict(V)

        # Provide each data point's anticipated cluster labels back.
        return y_predicted
```

# <h2>K-fold cross-validation with Silhouette score</h2>

```python
X = X
y_true = Y

cluster = []
# Taking range of total clusters
for i in range(2, 11):
    cluster.insert(1,i)

# silhouette score mean list
sc_mean = []
# silhouette score standard deviation list
sc_std = []

#Doing KFold cross-validation
k = 10
kf = KFold(n_splits=k)

for total_clusters in cluster:

    # Calling spectral clustering Class
    ssc = Scratch_SpectralClustering(total_clusters)
    y_predicted = ssc.fit_predict(X)

    # calculating the Silhouette Coefficient
    sc = []
    for train_set, test_set in kf.split(X):
        X_train, X_test = X[train_set], X[test_set]
        y_train, y_test = y_true[train_set], y_true[test_set]
        score = silhouette_score(X_test, y_test)
        sc.append(score)

    # Adding the mean silhouette score and standard deviation silhouette scores to each list.
    sc_mean.append(np.mean(sc))
    sc_std.append(np.std(sc))

# computing an average based on the mean silhouette scores and standard deviations silhouette scores
sc_mean_avg = np.mean(sc_mean)
sc_std_avg = np.mean(sc_std)

# Printing the values of average of mean and standard deviation of silhouette scores
print("Average of mean of Silhouette Coefficient is %.3f " %(sc_mean_avg))
print("Average of standard deviation of Silhouette Coefficient  is %.3f" %(sc_std_avg))
```

<h2>K-fold cross-validation with Normalized Mutual Information</h2>

```python
X = X
y_true = Y.ravel()

cluster = []
# Taking range of total clusters
for i in range(2, 11):
    cluster.insert(1,i)

# Normalized Mutual Information mean list
nmi_mean = []
# Normalized Mutual Information standard deviation list
nmi_std = []

#Doing KFold cross-validation
k = 10
kf = KFold(n_splits=k)

for total_clusters in cluster:
    # Calling spectral clustering Class
    ssc = Scratch_SpectralClustering(total_clusters)
    y_predicted = ssc.fit_predict(X)

    # calculating the Normalized Mutual Information
    nmi = []
    for train_set, test_set in kf.split(X):
        X_train, X_test = X[train_set], X[test_set]
        y_train, y_test = y_predicted[train_set], y_predicted[test_set]
        if len(y_test.shape) > 1:
            avg_y_test = np.mean(y_test)
        else:
            avg_y_test = y_test
        score = normalized_mutual_info_score(avg_y_test, y_true[test_set])
        nmi.append(score)

    # Adding the mean Normalized Mutual Information score and standard deviation Normalized Mutual Information scores to each list.
    nmi_mean.append(np.mean(nmi))
    nmi_std.append(np.std(nmi))

# computing an average based on the mean Normalized Mutual Information scores and standard deviations Normalized Mutual Information scores
nmi_mean_avg = np.mean(nmi_mean)
nmi_std_avg = np.mean(nmi_std)

# Printing the values of average of mean and standard deviation of Normalized Mutual Information scores
print("average of mean of Normalized Mutual Information score is %.3f" %(nmi_mean_avg))
print("average ofstandard deviation of Normalized Mutual Information score is %.3f" %(nmi_std_avg))
```
